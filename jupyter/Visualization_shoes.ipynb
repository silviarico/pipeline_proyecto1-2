```python
#en este kernel, visualizamos los datos para determinar más fácilmente si la marca nike a subido sus precios con respecto a 2018
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import re
import plotly.express as px
import plotly.graph_objects as go
```

```python
nike_co=pd.read_csv("../data_clean/nike_all.csv",encoding = "ISO-8859-1")
```

```python
nike_co.head()
```

```python
fig1=nike_co.groupby(["modelo", "an"])["prices.amountMax"].mean().unstack().plot(kind="bar",figsize=(15,8))
```

```python
#en el gráfico 1 podemos observar que, por lo general, en 2021 los precios de los modelos comparados son más altos que en 2018 a excepción los modelos de running lunar
```

```python
fig2 = px.box(nike_co, x="prices.amountMax", y="modelo", color="an", width=1000, height=600) #Points añade los puntos a la izquierda de cada box
fig2.show()
```

```python
#el gráfico 2, observamos que la mediana de precios es mucho más cercana que la media y que existe mayor dispersión entre los precios de 2021
```

```python
fig3 = px.scatter(nike_co, x="prices.amountMax", y="prices.amountMin", color="an")
fig3.show()
```

```python
#en el último gráfico podemos ver la relación entre el precio máximo(sin promoción) y el precio minímo(en promoción) de los productos de nike por fecha filtrada por año. 
#si bien la los datos de 2021 son más altos y muestra una correlación de precios que indica pocos descuentos, la muestra de datos de 2018 es demasiado pequeña para poder afirmar que Nike ha subido los precios. 
```
